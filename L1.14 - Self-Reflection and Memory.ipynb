```python
class Memory:
  def __init__(self):
      self.messages = []

  def add_message(self, role, content):
      self.messages.append({"role": role, "content": content})

  def get_messages(self):
      return self.messages

  def last_message(self):
      return self.messages[-1] if self.messages else None
```
